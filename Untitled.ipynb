```python
tr=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
en = inflect.engine()
def convert(label):
    word = en.number_to_words(label)
    term = [tr.index(i) for i in word]
    return term
    
    
```

```python
files =[]
labels = []
mfcc_features = []
PATH ="free-spoken-digit-dataset-master/recordings/"

for f in os.listdir(PATH):
    if(f.endswith(".wav")):
        label= f.split("_")[0]
        labels.append(convert(int(label)))
        file, sr = librosa.load(f"{PATH}{f}")
        mfcc = librosa.feature.mfcc(y=file, sr=sr)
        files.append(file)
        mfcc_features.append(mfcc)

```

```python
encoder = LSTM(50, return_state = True)
encoder_output, state_h, state_c = encoder(en_x)
encoder_states = [state_h, state_c]
```

```python
decoder_inputs = Input(shape=(None,))
dex = Embedding(26, 26)
final_dex = dex(decoder_inputs)

decoder_lstm = LSTM(50, return_sequences = True, return_state=True)

decoder_outputs , _ , _ = decoder_lstm(final_dex, initial_state=encoder_states)

decoder_dense = Dense(5, activation='softmax')

decoder_outputs = decoder_dense(decoder_outputs)

model = Model([encoder_input, decoder_inputs], decoder_outputs)


model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['acc'])


model.summary()
```
